```julia
function runInt!(prog)
    opcode = Dict(1 => +, 2 => *)
    pc = 1
    while prog[pc] != 99
        #println("Performing ", opcode[prog[pc]], " on ", prog[pc+1], " ", prog[pc+2])
        a, b, c = prog[pc+1:pc+3]
        prog[c+1] = opcode[prog[pc]](prog[a+1], prog[b+1])
        pc += 4
    end
    println("Done. Return value is ", prog[1])
    return prog[1]
end

function runInt(prog, a, b)
    prog[2] = a
    prog[3] = b
    return runInt(prog)
end

function runInt(prog)
    prog_copy = prog
    return runInt!(prog_copy)
end
```

```julia
proggie = open("input2.dat") do file
    input2 = read(file, String)
    [parse(Int, x) for x in split(input2, ',')]
end;
```

```julia
runInt(proggie, 12, 02)
```

```julia
function part2(prog)
    i = 0
    instrux = [(a,b) for a in 0:99 for b in 0:99];
    for (a,b) in instrux
        prog_copy = prog[:]
        x = runInt(prog_copy, a, b)
        if x == 19690720
            return 100 * a + b
        end
        i += 1
        println(i)
    end
    return 0
end
```
